```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Generate a stationary time series
np.random.seed(42)
num_points = 100
mean = 0
std_dev = 0
time = np.arange(0, num_points)
frequency = 0.1  # Adjust the frequency as needed

# stationary_series = np.random.normal(mean, std_dev, num_points)
stationary_series = np.sin(2 * np.pi * frequency * time)



# Plot the stationary time series
plt.figure(figsize=(10, 6))
plt.plot(stationary_series, marker='o')
plt.title("Stationary Time Series")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
```

```python
def test_stationarity(time_series):
    result = adfuller(time_series)
    p_value = result[1]
    significance_level = 0.05
    
    print(p_value)

    if p_value < significance_level:
        return True  # Series is stationary
    else:
        return False  # Series is not stationary
```
