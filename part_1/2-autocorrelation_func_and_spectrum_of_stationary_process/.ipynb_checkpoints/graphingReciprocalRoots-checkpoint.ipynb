# Graphing Reciprocal Roots

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to plot roots
def plot_roots(roots, title):
    plt.figure(figsize=(6, 6))
    plt.scatter(roots.real, roots.imag, color='red', label='Reciprocal Roots')
    unit_circle = plt.Circle((0, 0), 1, fill=False, color='blue', linestyle='dotted', label='Unit Circle')
    plt.gca().add_patch(unit_circle)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.title(title)
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.legend()
    plt.show()

# Close to One (Non-stationary)
roots_close_to_one = np.roots([1, -0.9, 0.81])
plot_roots(roots_close_to_one, 'Close to One (Non-stationary)')

# Greater Than One (Explosive Behavior)
roots_greater_than_one = np.roots([1, 0.5, 1])
plot_roots(roots_greater_than_one, 'Greater Than One (Explosive Behavior)')

# Less Than One (Damping and Stability)
roots_less_than_one = np.roots([1, -1.5, 0.7])
plot_roots(roots_less_than_one, 'Less Than One (Damping and Stability)')
```
