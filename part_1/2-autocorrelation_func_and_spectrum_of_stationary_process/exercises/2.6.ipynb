```python
def rename_col(df, unnamed_col, sales_col):
    """
    """
    
    new_df = df.rename(columns={unnamed_col:'Year', sales_col: 'Annual Sales'})
    
    return new_df
```

```python
sales_50_to_65_df = copy_mink_sales_df.loc[0:, ["Year", "Unnamed: 1"]]
unnamed_col = "Year"
sales_col = "Unnamed: 1"
updated_sales_50_to_65_df = rename_col(sales_50_to_65_df, unnamed_col, sales_col)
updated_sales_50_to_65_df
```

```python
sales_66_to_81_df = copy_mink_sales_df.loc[0:, ["Unnamed: 2", "Unnamed: 3"]]
unnamed_col = "Unnamed: 2"
sales_col = "Unnamed: 3"
updated_sales_66_to_81_df = rename_col(sales_66_to_81_df, unnamed_col, sales_col)
updated_sales_66_to_81_df
```

```python
sales_82_to_96_df = copy_mink_sales_df.loc[0:, ["Unnamed: 4", "Unnamed: 5"]]
unnamed_col = "Unnamed: 4"
sales_col = "Unnamed: 5"
updated_sales_82_to_96_df = rename_col(sales_82_to_96_df, unnamed_col, sales_col)
updated_sales_82_to_96_df
```

```python
sales_97_to_11_df = copy_mink_sales_df.loc[0:, ["Unnamed: 6", "Unnamed: 7"]]
unnamed_col = "Unnamed: 6"
sales_col = "Unnamed: 7"
updated_sales_97_to_11_df = rename_col(sales_97_to_11_df, unnamed_col, sales_col)
updated_sales_97_to_11_df
```

```python
mink_sales_df = pd.concat([updated_sales_50_to_65_df, updated_sales_66_to_81_df, updated_sales_82_to_96_df, updated_sales_97_to_11_df], ignore_index=True)
mink_sales_df.dropna(inplace=True)
```
