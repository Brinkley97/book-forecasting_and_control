# Calculate $\hat\beta$

```python
def beta(X, y):
    beta = (X * X.T)**(-1) * X.T * y
    beta_summed = beta.sum()
    return beta, beta_summed
```

# Intercept

- Based on https://medium.com/analytics-vidhya/ordinary-least-square-ols-method-for-linear-regression-ef8ca10aadfc


```python
def intercept(X, y, df_train_len):
    """
    
    Parameters:
    X -- pd Df
    y - pd Series
    
    Return
    intercept of b1_hat and b2_hat
    """
    
    y_mean = np.mean(y)
    y_errors = y - y_mean
    # print(y_mean, y_errors)

    X_mean = np.mean(X)
    X_errors = X - X_mean
    # print(X_mean, X_errors)
    
    b1_hat = np.sum(X_errors * y_errors) / np.sum(X_errors**2)
    # print(b1_hat)
    b0_hat = y_mean - (b1_hat * X_mean)
    # print(b0_hat)
    
    y_predictions = b0_hat + b1_hat * X
    # print(y_predictions)
    
    mse = ((y_predictions - y)**2).sum()
    # print(mse)
    # root_mse = mse**1/2
    root_mse = np.sqrt(mse/df_train_len)
    # print(root_mse)
    
    ssr = np.sum(y_errors**2)
    # print(ssr)
    
    # from : https://www.geeksforgeeks.org/ordinary-least-squares-ols-using-statsmodels/
    # try_ssr = np.sum(((y - b1_hat) * (X - b0_hat))**2)
    # print("try_ssr: ", try_ssr)
    
    sst = np.sum((y - y_predictions)**2)
    # print(sst)
    
    r_squared = sst / ssr
    # print(r_squared)
    
    score = 1 - r_squared
    # print(score) 
    
    # return 
    return b1_hat, b0_hat, mse, root_mse, r_squared, score
```
